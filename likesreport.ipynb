<h4>In order to help me compare these two tidied data frames, the next lines were written with the help of ChatGPT4</h4>

```python
common_index = likes_df3[likes_df3['title'].isin(followers_df4['value'])]
print("Common titles in likes_df3 that are also in followers_df4:", common_index)
```

```python
# Group by 'title', count the occurrences, and sort by 'value'
grouped_likes_df = likes_df3.groupby('title').count().sort_values('value', ascending=False)

# Filter to only include entries where the count of 'value' is 2 or more
filtered_likes_df = grouped_likes_df[grouped_likes_df['value'] >= 2]

# Display the top 2000 rows
top_2000_filtered_likes_df = filtered_likes_df.head(2000)

print(top_2000_filtered_likes_df)
```

<h5>Above is account filtered to show only those with two or more liked posts from me</h5>
